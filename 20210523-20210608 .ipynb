```python
import geopandas as gpd
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import timedelta
```

```python
gdf = pd.read_csv('/Users/wudailing/Desktop/US Ignite/trips_after_cleaning_old.csv')
```

```python
trips_0523_to_0608 = gdf[(gdf['MST_start_datetime']>='2021-05-23')&\
                             (gdf['MST_start_datetime']<='2021-06-08')]
print('total number of trips between 20210523 and 20210608:{:,}'.format(len(trips_0523_to_0608)))
print('geospatial trip breakdown:')
trips_0523_to_0608.groupby(by = ["GeospatialType"]).count()['TripId'].apply(lambda x : "{:,}".format(x))
```

```python
ACP_sensor = 59922+5469+113988+71043+83715+24647
EI = trips_0523_to_0608.groupby(by = ["GeospatialType"]).count()['TripId']['EI']
trip_share = EI/ACP_sensor

print('ACP sensor:{:,}'.format(ACP_sensor))
print('trips onto base sensor:{:,}'.format(EI))
print('trip share:{:.1%}'.format(trip_share))
```
